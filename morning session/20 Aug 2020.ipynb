### Data visualization
- data visualization is the reprsentation of data in graphical format
- huge amount of data  is represented in a simple and easy to understand format
- import matplotlib.pyplot

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
plt.plot([1,2,3],[1,2,3],'go')
plt.show()
```

```python
plt.title("Linear graph with dashed lines")
plt.plot([1,2,3],[1,2,3],'r--')
plt.show()
```

```python
plt.title("Linear graph with dashed lines")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot([1,2,3],[1,2,3],'k.') # ':' -- dotted line,'s'-- squared , '^'-- triangle, '-' --- linear line
plt.axis([0,4,0,4])
plt.show()
```

```python
student = ['archana','vidya','lalitha']
marks = [10,15,20]

plt.figure(figsize=(9,3))

plt.subplot(131) # (111)
plt.bar(student,marks,color='violet')

plt.subplot(132)
plt.scatter(student,marks,color = 'green')

plt.subplot(133)
plt.plot(student,marks,'k^')
plt.suptitle("categorical plotting")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(1)                
plt.subplot(211)             
plt.plot([1, 2, 3])
plt.subplot(212)             
plt.plot([4, 5, 6])

plt.figure(2)                
plt.plot([4,5,6])
```

```python
activities = 'eating','sleeping','playing','reading'
sizes = [25,50,10,15]
explode = (0,0,0.1,0)
fig1,ax1 = plt.subplots()

ax1.pie(sizes,explode=explode,labels = activities,autopct = '%1.1f%%',
        shadow = True,startangle=90)
plt.show()
```

```python
a = np.array([22,15,16,25,38,45,65,78])
plt.hist(a,bins = [0,25,50,75,100])
plt.show()
```
